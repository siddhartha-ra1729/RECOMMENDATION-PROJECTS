```python
num_rating = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating
```

```python
popular = num_rating.merge(avg_rating,on='Book-Title')
popular
```

```python
popular = popular[popular['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
similarity_score
```

```python
print(pt.index.tolist())  # This will give me a list of all book titles
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        book_title = pt.index[i[0]]
        print(book_title)
```

```python
def recommend(book_name):
    try:
        index = pt.index.get_loc(book_name)
    except KeyError:
        print(f"Book '{book_name}' not found in the index.")
        return
    
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    for i in similar_items:
        book_title = pt.index[i[0]]
        print(book_title)

recommend('1984')
```

```python
def recommend(book_name):
    if book_name not in pt.index:
        print(f"Book '{book_name}' not found in the index.")
        return
    
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    for i in similar_items:
        book_title = pt.index[i[0]]
        print(book_title)

recommend('The Bad Place')
```

# similarity_score: This variable is expected to be a 2D array (like a matrix) that holds the similarity scores between books. Each row corresponds to a book, and the values in that row represent the similarity to other books

# The recommend function suggests similar books based on a given title. It checks if the book exists, retrieves its index, calculates similarity scores from a matrix, and prints the titles of the top 5 most similar books, excluding the original book itself.

```python
recommend('Fried Green Tomatoes at the Whistle Stop Cafe')
```

# Visualization 1: Most Prolific Authors

```python
plt.figure(figsize=(12, 6))
top_authors = book['Book-Author'].value_counts().head(10)
sns.barplot(x=top_authors.values, y=top_authors.index, palette='viridis')
plt.title('Top 10 Most Prolific Authors')
plt.xlabel('Number of Books')
plt.ylabel('Authors')
plt.show()
```

```python
# Top Publishers
plt.figure(figsize=(12, 6))
top_publishers = book['Publisher'].value_counts().head(10)
sns.barplot(x=top_publishers.values, y=top_publishers.index, palette='magma')
plt.title('Top 10 Publishers')
plt.xlabel('Number of Books')
plt.ylabel('Publishers')
plt.show()
```

```python
# Count of Books by Year
plt.figure(figsize=(12, 6))
sns.countplot(data=book, x='Year-Of-Publication', palette='crest', order=book['Year-Of-Publication'].value_counts().index)
plt.title('Count of Books by Year')
plt.xlabel('Year of Publication')
plt.ylabel('Number of Books')
plt.xticks(rotation=45)
plt.show()
```
